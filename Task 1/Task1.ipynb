```python
import pandas as pd
df = pd.read_csv(
    r"C:\Users\Bhadmus\Documents\online_retail_II.xlsx.csv",
    encoding='ISO-8859-1'
)
df
```

```python
#Missing values
df.isnull().sum() #this shows the number of missing values in each column. There are about 135080 missing values in the Customer ID column and about 1454 missing values in the Description column.b
df['Customer ID'] = df['Customer ID'].fillna(0) #there were about 135080 missing values in the Customer ID column, I have filled them with 0 since it is a numeric column and it will not affect the analysis.
df['Customer ID'] = df['Customer ID'].astype(int)
df['Description'] = df['Description'].fillna('Unknown') #there were about 1454 missing values in the Description column, I have filled them with 'Unknown' since it is a categorical column and it will not affect the analysis.
df.isnull().sum()
df
```

```python
df['InvoiceDate']= df['InvoiceDate'].astype(str).str.strip() #this will remove any leading or trailing spaces from the InvoiceDate column and convert it to string type.
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'],dayfirst=True, format='mixed', errors='coerce') #this will convert the InvoiceDate column to datetime format. If there are any invalid date formats, they will be set to NaT (Not a Time).
df['InvoiceDate'].isnull().sum() #this will show the number of missing values
df['InvoiceDate'].dtype #this will show the data type of the InvoiceDate column, which should be datetime64[ns] after the conversion.
df
```

```python
df['Description'] = df['Description'].str.strip()
df['Description'] = df['Description'].str.lower()
df['Description'] = df['Description'].str.replace(' ', '_')
df['Country'] = df['Country'].str.strip()
df.duplicated().sum() #to check if there are any duplicated rows in the dataset after cleaning the Description and Country columns. This shows that there are no duplicated rows in the dataset.
# This shows EVERY row that has a twin somewhere else in the data
duplicates_to_inspect = df[df.duplicated(keep=False)].sort_values(by='Invoice')

# Display the first 10
duplicates_to_inspect

```

```python
df.drop_duplicates(inplace=True)#to remove any duplicated rows in the dataset after cleaning the Description and Country columns. This will ensure that there are no duplicated rows in the dataset
df['Revenue'] = df['Quantity'] * df['Price'] #this creates a new column called Revenue which is the product of Quantity and Price columns. This will be useful for analysis later on.
df
```

```python
df_customers = df[(df['Customer ID'] != 0) & (df['Description'] != 'Unknown')] #this creates a new dataframe called df_customers which only contains rows where the CustomerID is not 0 and the Description is not 'Unknown'. This will be useful for analysis later on since we want to focus on customers who have made purchases and have valid descriptions.
df_customers
```

```python
df_customers.columns = (df_customers
                        .columns
                      .str.strip() 
                      .str.replace(' ', '_')
                      ) #this cleans the column names by stripping any leading or trailing whitespace, converting them to lowercase, and replacing any spaces with underscores. This will make it easier to work with the column names in the analysis later on.
```
